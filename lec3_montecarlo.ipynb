```python
#monte carlo integration
import numpy as np

xmax=10000
xmin=0

xi=np.random.uniform(xmin,xmax, 100000)
sigma=7
def f(x):
    return (x**3)*np.exp(-(x**2)/(2*(sigma**2)))

integral=(xmax-xmin)*np.mean(f(xi))

result=2*sigma**4

print(integral)
print(result)


#does the result converge with the number of samples?
#how does the error go down?

#do it many times
#for a given N how are the result distributed?
#how does the distribution change if N increases?
```
