```python
import random
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#a function that simulates the monty hall "experimen"
def montyhall(trial):
    #counter for winnings of the player stays or switches
    stay=0
    switch= 0

    for i in range(trial):
        prize = random.randint(1,3)
        player_choice = random.randint(1,3)

        possible_doors = [1,2,3]
        possible_doors.remove(player_choice) 
        #the host only opens doors that doesnt have the price in it 
        if prize in possible_doors:
            possible_doors.remove(prize)

        host_opens= random.choice(possible_doors)
        #the remainder doors...
        remaining_doors = [1,2,3]
        remaining_doors.remove(player_choice)
        remaining_doors.remove(host_opens)
        switch_choice= remaining_doors[0]


        if player_choice==prize:
            stay += 1


        if switch_choice ==prize:
            switch += 1
    return stay,switch






```

```python
trials = 10000
stay, switch = montyhall(trials)

print("Trials:", trials)
print("Stay wins:", stay)
print("Switch wins:", switch)
print("The probablity of staying with your choice and winning is", stay/trials*100,"%")
print("The probablity of switching your choice and winning is", switch/trials*100,"%")
```
