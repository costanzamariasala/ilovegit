```python
#monte carlo integration
import numpy as np

sigma=1 #parameter

N=1000 #number of samples

#random samples from Gaussian
x=np.random.normal(0,sigma, N) #0 is the center of the distirbution

#positive values 
x=x[x>0]

#monte carlo integral
integral=sigma*np.sqrt(np.pi/2) *np.mean(x**3)

print('Monte Carlo integral:', integral)
print('Exact value:', 2*sigma**)




#does the result converge with the number of samples?
#how does the error go down?

#do it many times
#for a given N how are the result distributed?
#how does the distribution change if N increases?
```
